```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('creditcard.csv')
df.shape
```

Since the data size is very large(nearly 300k) we will use only a fraction of data which can still give us very high accuracy because we just need to make binary classification

```python
df = df.sample(frac=0.25,random_state=27)
df.shape
```

```python
df.hist(figsize=(20,20))
plt.show()
```

```python
(df['Class'].value_counts()/len(df)).plot.bar()
print(df['Class'].value_counts()/len(df))
```

Only 0.18% of all transactions result in a fraud

```python
# making co-relation matrix for better understanding of relations b/w the data
CoMatrix = df.corr()
import seaborn
plt.figure(figsize=(7,5),dpi=120)
seaborn.heatmap(CoMatrix)
```

Since columns like V14,V12,V16,V15,V17,V18,V7,V5 have negative co-relation with target variable we need to consider them in our prediction model

```python
df.drop(['V14','V12','V16','V15','V17','V18','V7','V5'],axis=1,inplace=True)
```
